Напишите функцию, которая будет принимать в качестве аргумента букву и выводить все слова из строки, начинающиеся на эту букву (например, “w”).

```python
import re

def get_words_begin_with(some_string, letter):
    '''
    Возвращает список слов в строке, начинающихся на указанную букву.
    Args:
        some_string (str): Исходный текст.
        letter (str): Буква для поиска (в нижнем регистре).
    Returns:
        list: Список слов, начинающихся на letter.
    '''
    words = []
    for word in re.findall(r'\b\w+\b', some_string.lower()):
        if word[0] == letter:
            words.append(word)
    return words

def letter_input(some_string):
    '''Запрашивает букву у пользователя и выводит все слова из текста,
    начинающиеся на эту букву.'''
    letter = input('введите букву ')

    words = get_words_begin_with(some_string, letter)

    for word in words:
        print(word)
```

#### Задача 3

Напишите функцию, которая генерирует случайную последовательность из 6 цифр, причем в этой последовательности должна быть как минимум одна цифра 3. Позиция 3 определяться случайным образом. Например: "456379" или "033456".

```python
import random
```

```python
def get_rnd_6_dig_sequence_with_3():
    '''
    Генерирует 6-значную строку, содержащую минимум одну цифру '3'.

    Returns:
        str: Шестизначная последовательность цифр (возможны ведущие нули).
    '''
    sequence = [str(random.randint(0, 9)) for _ in range(6)]
    # получение случайного индекса от 0 до 5 и вставка в него тройки
    position = random.randint(0,5)
    sequence[position] = '3'
    return ''.join(sequence)

get_rnd_6_dig_sequence_with_3()
```

1. Преобразуйте данные таким образом, чтобы они были представлены в виде словаря, в котором ключами являются номера студентов, а значениями — список из других данных о студентах.

```python
students_dict = {}

for student in students:
    student_id = student[0]
    student_info = student[1:]
    students_dict.setdefault(student_id, [])
    students_dict[student_id] += student_info

students_dict
```

2. Напишите функцию, которая принимает номер группы и выводит пронумерованный и упорядоченный по алфавиту список (фамилии, имена, отчества) всех студентов этой группы. Необходимо обращаться к словарю.

```python
def get_students_list_in_group(group_number):

    students_list = []

    for value in students_dict.values():
        if value[-1] == group_number:
            fio = ' '.join(value[:3])
            students_list.append(fio)

    students_list = sorted(students_list)

    result = []
    for i, fio in enumerate(students_list, start=1):
        result.append(f'{i} {fio}')

    return '\n'.join(result)
```

```python
print(get_students_list_in_group('БСТ161'))
```

#### Задача 6

Дана матрица matrix. Напишите код, который будет транспонировать её (то есть, делать из строк столбцы) в maytix_t:

matrix_t = [
    [0, 6, 3, 4, 5],
    [1, 2, 3, 6, 7],
    [2, 2, 3, 7, 3],
    [4, 1, 3, 1, 4],
    [8, 9, 3, 2, 0]
]

```python
matrix = [
    [0,1,2,4,8],
    [6,2,2,1,9],
    [3,3,3,3,3],
    [4,6,7,1,2],
    [5,7,3,4,0]
]
```

```python
# определяем количество строк rows и столбцов cols в исходной матрице
rows = len(matrix)
cols = len(matrix[0])

'''
[0] * rows	Создаёт список из rows нулей (одна строка)
for _ in range(cols)	Повторяет действие cols раз
[...]	Собирает всё в один список (матрицу)
Результат: матрица размером cols × rows, заполненная нулями.
То есть получаем новую матрицу с транспонированными размерами от исходной:
rows столбцов
cols строк
'''
matrix_t = [[0] * rows for _ in range(cols)]

# вложенный цикл заполняет новую матрицу, транспонируя исходную
for i in range(cols):
    for j in range(rows):
        matrix_t[i][j] = matrix[j][i]
print(matrix_t)
```
