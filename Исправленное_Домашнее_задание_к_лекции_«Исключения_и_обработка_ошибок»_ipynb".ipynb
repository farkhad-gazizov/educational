**Задание 2**

Дан поток дат в формате YYYY-MM-DD, в которых встречаются некорректные значения:

`stream = [‘2018-04-02’, ‘2018-02-29’, ‘2018-19-02’]`

Напишите функцию, которая проверяет эти даты на корректность. То есть для каждой даты возвращает True — дата корректна или False — некорректная.

````python
from datetime import datetime

def date_is_correct(date_string):
    '''
    Проверяет, является ли строка корректной датой в формате YYYY-MM-DD.
    Args:
        date_string (str): Дата в виде строки.
    Returns:
        bool: True, если дата корректна, иначе False.
    '''
    try:
        datetime.strptime(date_string, '%Y-%m-%d')
        return True
    except ValueError:
        return False

# используем функцию в цикле для заданного потока
stream = ['2018-04-02', '2018-02-29', '2018-19-02']

for date_string in stream:
    print(f'{date_string} - {date_is_correct(date_string)}')
````

**Задание 3**

Напишите функцию `date_range`, которая возвращает список дат за период от `start_date` до `end_date`. Даты должны вводиться в формате `YYYY-MM-DD`. В случае неверного формата или при `start_date > end_date` должен возвращаться пустой список.

````python
from datetime import datetime, timedelta

def date_range(start_date, end_date):
    '''
    Возвращает список дат от start_date до end_date включительно
    в формате YYYY-MM-DD.
    Args:
        start_date (str): Начальная дата в формате YYYY-MM-DD.
        end_date (str): Конечная дата в формате YYYY-MM-DD.
    Returns:
        list: Список строк с датами. При неверном формате любой из дат
              или при start_date > end_date возвращает пустой список.
    '''
    # проверка формата стартовой даты
    try:
        start_date_dt = datetime.strptime(start_date, '%Y-%m-%d')
    except ValueError:
        return []

    # проверка формата конечной даты
    try:
        end_date_dt = datetime.strptime(end_date, '%Y-%m-%d')
    except ValueError:
        return []

    # проверка случая start_date > end_date
    if start_date_dt > end_date_dt:
        return []

    # формирование списка дат в случае, если все корректно
    dates = []
    current_date_dt = start_date_dt
    while current_date_dt <= end_date_dt:
        dates.append(current_date_dt.strftime('%Y-%m-%d'))
        current_date_dt += timedelta(days=1)

    return dates
````

````python
# проверка работы функции date_range на тестовых данных
# пограничные даты корректны, стартовая не больше конечной
start_date = '2026-05-01'
end_date = '2026-05-09'

print(date_range(start_date, end_date))
````

````python
# проверка работы функции date_range на тестовых данных
# одна из пограничных дат имеет некорректный формат
start_date = '2026-05-01'
end_date = '2026-05-33'

print(date_range(start_date, end_date))
````

````python
# проверка работы функции date_range на тестовых данных
# стартовая дата больше конечной
start_date = '2026-05-01'
end_date = '2026-04-30'

print(date_range(start_date, end_date))
````

Ваш коллега прислал код функции:

```
DEFAULT_USER_COUNT = 3

def delete_and_return_last_user(region, default_list=[‘A100’, ‘A101’, ‘A102’]):

```




При однократном вызове этой функции всё работает корректно:
`delete_and_return_last_user(1)`
`‘A101’`

Однако при повторном вызове получается ошибка `IndexError: list index out of range.`

Задание:

Что значит ошибка list index out of range?
Почему при первом запуске функция работает корректно, а при втором — нет?

**Ответ.**

Ошибка `list index out of range` означает, что происходит обращение к элементу списка по индексу, которого не существует. При первом вызове функции в списке еще есть элемент с индексом 1, это 'A101':

 `default_list[DEFAULT_USER_COUNT-2]`, что равняется `default_list[3-2]`,

поэтому функция работает корректно. К моменту второго вызова в списке остается меньше 2х элементов, и обращение к элементу с индексом 1 приводит к указанной ошибке. То есть, после первого вызова в памяти функции остается измененный список `['A100', 'A101']`, во время второго вызова удаляется элемент 'A101' и список становится `['A100']`, в котором уже нет элемента с индексом 1. Первопричина ошибки в том, что в качестве значения функции по умолчанию используется изменяемый объект - список.

**Корректное решение:**

````python
DEFAULT_USER_COUNT = 3

def delete_and_return_last_user(region, default_list=None):
    '''
    Удаляет из списка default_list последнего пользователя
    и возвращает ID нового последнего пользователя.
    '''
    if default_list is None:
        default_list=['A100', 'A101', 'A102']

    element_to_delete = default_list[-1]
    default_list.remove(element_to_delete)

    return default_list[DEFAULT_USER_COUNT-2]

delete_and_return_last_user(1)
````
