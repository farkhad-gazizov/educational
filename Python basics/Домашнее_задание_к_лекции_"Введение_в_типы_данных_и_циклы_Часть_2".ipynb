## Задание 1
Дана переменная, в которой хранится словарь, содержащий гео-метки для каждого пользователя (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран множество уникальных гео-меток всех пользователей.

Пример работы программы:  

```
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}
```

Результат:
`{98, 35, 15, 213, 54, 119}`

````python
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}

'''
переводим во множество (для получения уникальных значений) вновь создаваемый
список из гео-меток (с пом. list comprehension) - дословно берем и кладем в
список метки (geo) из каждого списка (внутренний цикл - правая часть выражения),
перебирая каждое значение исходного словаря (внешний цикл - левая часть выражения)
'''
result = set([geo for values in ids.values() for geo in values])
result
````

## Задание 2
Дана переменная, в которой хранится список поисковых запросов пользователя (пример структуры данных приведен ниже, но запросы потенциально могут быть любые). Вам необходимо написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

Пример работы программы:  

```
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
```

Результат:
```
Поисковых запросов, содержащих 2 слов(а): 42.86%
Поисковых запросов, содержащих 3 слов(а): 57.14%
```

````python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

queries_dict = {}

for query in queries:
    word_count = len(query.split())
    queries_dict.setdefault(word_count, 0)
    queries_dict[word_count] += 1

for key, value in sorted(queries_dict.items()):
    print(f'Поисковых запросов, содержащих {key} слов(а): {(value / len(queries)):.2%}')
````

## Задание 5 (необязательно)
Дан список произвольной длины. Необходимо написать код, который на основе исходного списка составит словарь такого уровня вложенности, какова длина исхондого списка.

Примеры работы программы:  

1. `my_list = ['2018-01-01', 'yandex', 'cpc', 100]`

Результат:
`{'2018-01-01': {'yandex': {'cpc': 100}}}`

2. `my_list = ['a', 'b', 'c', 'd', 'e', 'f']`

Результат:
`{'a': {'b': {'c': {'d': {'e': 'f'}}}}}`

````python
my_list = ['a', 'b', 'c', 'd', 'e', 'f']

# защита от неподходящего случая - при количестве элементов списка менее 2
if len(my_list) < 2:
    print(my_list)
else:
# принимаем последний элемент списка в качестве значения самого глубокого
# вложения создаваемого списка
    my_dict = my_list[-1]

# в цикле перебираем список, создавая на каждой итерации словарь с вложениями
# таким образом, что предпоследний элемент списка - ключ, последний - значение
    for i in range(len(my_list) - 2, -1, -1):
        my_dict = {my_list[i]:my_dict}

    print(my_dict)
````

Результат:

````python
cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},
    ]
}

serving_number = int(input('Укажите количество порций: '))

total_ingredients = {}

'''
из каждой строки (отдельный словарь) рецепта (отдельный список) каждого
блюда собираем новый словарь total_ingredients, где ключи - кортеж из названия
ингредиента и его меры (сделано так, потому что у одного ингредиента могут быть
разные меры), а значения - количество. setdefault при отсутствии в новом словаре
ключа добавляет его со значением ноль, а потом суммирует значения, при наличии -
ничего не делает. Затем циклом и с f-строкой выводим ключи-значения нового
словаря согласно условию задачи
'''
for recipe in cook_book.values():
    for element in recipe:
        ingredient_measure = tuple([element['ingridient_name'], element['measure']])
        quantity = element['quantity'] * serving_number
        total_ingredients.setdefault(ingredient_measure, 0)
        total_ingredients[ingredient_measure] += quantity

for key, value in total_ingredients.items():
    print(f'{key[0].capitalize()}: {value} {key[1]}')
````
