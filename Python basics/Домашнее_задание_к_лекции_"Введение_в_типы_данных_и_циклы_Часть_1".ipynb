**Задание 1.**

Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

*   среднюю букву, если число букв в слове нечетное;
*   две средних буквы, если число букв четное.


```python
word = input('Введите слово: ')
mid_of_word = len(word) // 2
if len(word) % 2 == 0:
    first_half_of_word = word[:mid_of_word]
    second_half_of_word = word[mid_of_word:]
    print(first_half_of_word[-1], second_half_of_word[0], sep='')
else:
    print(word[mid_of_word])
```

**Задание 2**.

Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

```python

total = 0
num = float(input('введите число: '))
while num != 0:
    total += num
    num = float(input('введите число: '))
print(total)
```

**Задание 3.**

Мы делаем MVP dating-сервиса, и у нас есть список парней и девушек.
Выдвигаем гипотезу: лучшие рекомендации мы получим, если просто отсортируем имена по алфавиту и познакомим людей с одинаковыми индексами после сортировки! Но мы не будем никого знакомить, если кто-то может остаться без пары!

```python
boys = input('Введите имена парней через пробел: ').split()
girls = input('Введите имена девушек через пробел: ').split()
boys_sorted = sorted(boys)
girls_sorted = sorted(girls)

# при равном количестве парней и девушек образуем новый список, в котором они соединены попарно по одинаковым индексам
if len(boys_sorted) == len(girls_sorted):
    perfect_couples = zip(boys_sorted, girls_sorted)
    perfect_couples = list(perfect_couples)
    print('Идеальные пары:')
    for couples in perfect_couples:
        print(couples[0], 'и', couples[1])
else:
    print('Внимание, кто-то может остаться без пары!')
```

**Задание 4.**

У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере). Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.

```python
countries_temperature = [
    ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Германия', [57.2, 55.4, 59, 59, 53.6]],
    ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]

countries_average_celsius_temperature = []

for row in countries_temperature:
    # получаем список стран из первоначального списка
    country = row[0]
    # получаем список температур для каждой страны
    fahrenheit_temps = row[1]
    # перевод фаренгейтов в цельсии
    cels_temps = [(fahr_temp - 32) * 5/9 for fahr_temp in fahrenheit_temps]
    # вычисление средней температуры в цельсиях
    average_celsius_temperature = round(sum(cels_temps) / len(cels_temps), 1)
    # заполнение данными списка "страна - средняя температура в цельсиях"
    countries_average_celsius_temperature.append([country, average_celsius_temperature])

print('Средняя температура в странах:')

for data in countries_average_celsius_temperature:
    print(data[0], '-', data[1], 'C')
```

**Задание 5** (необязательное).

Имеется список с транспортными номерами. Необходимо написать код, который проверит каждый номер на валидность (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то вывести его в нужном формавте (пример ниже), а если не валиден — вывести текст. При решении помогут регулярные выражения, с которыми будет знакомство на практике.

```python
import re
car_ids = ['А222ВС96', 'АБ22ВВ193']
# создаем шаблон российских автомобильных номеров с используемыми буквами
# кириллицы, со ограничением начала и конца строки
car_ids_pattern = r'^[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}$'

for id_ in car_ids:
    # проверка гос.номеров в списке на соответствие шаблону
    if re.match(car_ids_pattern, id_):
        # разделяем номерную часть и регион в зависимости от длины гос.номера
        if len(id_) == 8:
            region = id_[-2:]
            number = id_[:-2]
        else:
            region = id_[-3:]
            number = id_[:-3]

        print(f'Номер {number} валиден. Регион: {region}')
    else:
        print(f'Номер {id_} невалиден')
```

**Задание 6** (необязательное).

Дан поток логов по количеству просмотренных страниц для каждого пользователя (пользователь,дата;просмотры). Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей. И тут регулярные выражения облегчат немного реализацию.

```python
import re
stream = [
    'user4,2021-01-01;3',
    'user3,2022-01-07;4',
    'user2,2022-03-29;1',
    'user1,2020-04-04;13',
    'user2,2022-01-05;7',
    'user1,2021-06-14;4',
    'user3,2022-07-02;10',
    'user4,2021-03-21;19',
    'user4,2022-03-22;4',
    'user4,2022-04-22;8',
    'user4,2021-05-03;9',
    'user4,2022-05-11;11'
]

user_list = []
views_list = []

for row in stream:
    splitted_row = re.split('[,;]', row)  # разделяем записи в списке логов по указанным разделителям
    if splitted_row[0] not in user_list:  # заполняем список уникальных пользователей
        user_list.append(splitted_row[0])
    views_list.append(splitted_row[2])  # заполняем список всех просмотров

views_list = [int(x) for x in views_list]  # преобразование в числовой тип данных
result = round(sum(views_list) / len(user_list), 2)
print('Среднее количество просмотров на уникального пользователя: ', result)
```
