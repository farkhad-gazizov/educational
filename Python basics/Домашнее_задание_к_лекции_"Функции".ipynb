# Домашнее задание к лекции "Функции"

Представьте, что вы помогаете библиотекарю автоматизировать учёт книг. Вам нужно написать программу, которая обрабатывает пользовательские команды на основе имеющихся данных.

Исходные данные имеют следующую структуру:

**1. Перечень всех книг**
```
books = [
    {'genre': 'поэзия', 'number': '978-5-1000-1234-7', 'title': 'Евгений Онегин', 'author': 'Александр Пушкин'},
    {'genre': 'фэнтези', 'number': '88006', 'title': 'Властелин колец', 'author': 'Джон Р. Р. Толкин'},
    {'genre': 'детектив', 'number': 'D-1122', 'title': 'Безмолвный свидетель', 'author': 'Агата Кристи'}
]
```

*   genre — жанр книги
*   number — уникальный номер
*   title — название книги
*   author — автор книги

**2. Перечень полок, на которых хранятся книги** (если книга есть в books, то она обязательно должна быть и в directories)
```
directories = {
    '1': ['978-5-1000-1234-7', '88006'],
    '2': ['D-1122'],
    '3': []
}
```


*   ключ словаря ('1', '2', '3') — это номер полки
*   значение — список номеров книг, лежащих на соответствующей полке



**Общие требования к программе:**
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **books** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**");
- код обязательно должен содержать конструкцию **match case** и функцию **setdefault()**.

````python
# Исходные данные: перечень книг (books) и перечень полок, на которых они находятся (directories)

books = [
    {'genre': 'поэзия', 'number': '978-5-1000-1234-7', 'title': 'Евгений Онегин', 'author': 'Александр Пушкин'},
    {'genre': 'фэнтези', 'number': '88006', 'title': 'Властелин колец', 'author': 'Джон Р. Р. Толкин'},
    {'genre': 'детектив', 'number': 'D-1122', 'title': 'Безмолвный свидетель', 'author': 'Агата Кристи'}
]

directories = {
    '1': ['978-5-1000-1234-7', '88006'],
    '2': ['D-1122'],
    '3': []
}
````

````python
def get_shelves_by_book_number(book_number, directories):
    '''
    Функция принимает номер книги в качестве аргумента,
    ищет на какой полке она лежит и возвращает номер полки.
    Если книга нигде не лежит, возвращает None

    '''
    for shelf, books_on_shelf in directories.items():
        if book_number in books_on_shelf:
            return shelf
    return None
````

````python
def shelves_list():
    """
    Возвращает строку с отсортированным по возрастанию списком номеров полок.
    Номера разделяются запятой
    """
    return ", ".join(sorted(directories.keys(), key=int))
````

### Пункт 1. Пользователь по команде "*book_info*" может найти название книги и её автора

````python
def find_book_info(book_number, books):
    """
    Находит книгу по её номеру и возвращает информацию о ней.
    Args:
        book_number (str): Уникальный номер книги.
        books (list): Список словарей с данными о книгах.
    Returns:
        str: Строка с названием и автором книги или сообщение,
        если книга не найдена.
    """
    for book in books:
        if book['number'] == book_number:
            return f"Название книги: {book['title']}\nАвтор: {book['author']}"
    return 'Книга не найдена в базе'

def book_info(books):
    """
    Запрашивает у пользователя номер книги и выводит информацию о ней.
    Args:
        books (list): Список словарей с данными о книгах.
    Returns:
        None
    """
    book_number = input('Введите номер книги:\n')
    print(find_book_info(book_number, books))
````

### Пункт 2. Пользователь по команде "*shelf*" может по названию книги узнать, на какой полке она хранится

Примеры работы:

1.

```
Введите команду:

Введите название книги:

```
Результат:

2.
```
Введите команду:

Введите название книги:
Абвгд
```
Результат:  
`Книга не найдена в базе`

````python
def find_shelf(book_title, books, directories):
    """
    Находит полку, на которой хранится книга по её названию.
    Args:
        book_title (str): Название книги.
        books (list): Список словарей с данными о книгах.
        directories (dict): Словарь с полками и номерами книг.
    Returns:
        str: Строка с номером полки или сообщение,
        если книга не найдена.
    """
    book_number = None
    for book in books:
        if book['title'].lower() == book_title.lower():
            book_number = book['number']
            break

    if book_number:
        shelf = get_shelves_by_book_number(book_number, directories)
        if shelf:
            return f"Книга хранится на полке: {shelf}"
    return 'Книга не найдена в базе'

def shelf(books, directories):
    """
    Запрашивает у пользователя название книги и выводит номер полки.
    Args:
        books (list): Список словарей с данными о книгах.
        directories (dict): Словарь с полками и номерами книг.
    Returns:
        None
    """
    book_title = input('Введите название книги:\n')
    print(find_shelf(book_title, books, directories))
````

### Пункт 3. Пользователь по команде "*all*" может увидеть полную информацию по всем книгам

Пример работы:

Результат:

````python
def all_(books, directories):
    """
    Выводит полную информацию по всем книгам, включая номер полки.
    Args:
        books (list): Список словарей с данными о книгах.
        directories (dict): Словарь с полками и номерами книг.
    Returns:
        None
    """
    for book in books:
        book_number = book['number']
        #shelf_number = None
        shelf_number = get_shelves_by_book_number(book_number, directories)
        #break
        print(f"№: {book['number']}, жанр: {book['genre']}, название: {book['title']}, автор: {book['author']}, полка хранения: {shelf_number}")
````

### Пункт 4. Пользователь по команде "*add_shelf*" может добавить новую полку

Примеры работы:

1.

```
Введите команду:
add_shelf

Введите номер полки:
10
```
Результат:  
`Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.`

2.
```
Введите команду:
add_shelf

Введите номер полки:
1
```
Результат:  
`Такая полка уже существует. Текущий перечень полок: 1, 2, 3.`

````python
def add_shelf(directories):
    """
    Добавляет новую полку в словарь полок по запросу пользователя.
    Если полка с указанным номером уже существует, новая полка не добавляется,
    и выводится соответствующее сообщение.
    Args:
        directories (dict): Словарь с полками и списками номеров книг.
    Returns:
        None
    """
    new_shelf = input('Введите номер полки:\n')
    if new_shelf in directories:
        print(f"Такая полка уже существует. Текущий перечень полок: {shelves_list()}.")
    else:
        directories.setdefault(new_shelf, [])
        print(f"Полка добавлена. Текущий перечень полок: {shelves_list()}.")
````

### Пункт 5. Пользователь по команде "*del_shelf*" может удалить существующую полку из данных (только если она пустая)

Примеры работы:

1.

```
Введите команду:

Введите номер полки:

```
Результат:  
`Полка удалена. Текущий перечень полок: 1, 2.`


2.
```
Введите команду:

Введите номер полки:

```
Результат:  
`На полке есть книги, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.`


3.
```
Введите команду:
del_shelf

Введите номер полки:
4
```
Результат:  
`Такой полки не существует. Текущий перечень полок: 1, 2, 3.`

````python
def del_shelf(directories):
    """
    Удаляет полку из словаря по запросу пользователя.
    Полка удаляется только если:
    - она существует в словаре;
    - на ней нет книг.
    Если полка не существует или содержит книги, удаление не выполняется,
    и выводится соответствующее сообщение.
    Args:
        directories (dict): Словарь с полками и списками номеров книг.
    Returns:
        None
    """
    delete_shelf = input('Введите номер полки:\n')
    if delete_shelf in directories:
        if not directories[delete_shelf]:
            del(directories[delete_shelf])
            print(f"Полка удалена. Текущий перечень полок: {shelves_list()}.")
        else:
            print(f"На полке есть книги, удалите их перед удалением полки. Текущий перечень полок: {shelves_list()}.")

    else:
        print(f"Такой полки не существует. Текущий перечень полок: {shelves_list()}.")
````

№: D-1122, жанр: детектив, название: Безмолвный свидетель, автор: Агата Кристи, полка хранения: 2

````python
def add_new_book_to_library(
    books,
    directories,
    new_book_number,
    new_book_genre,
    new_book_title,
    new_book_author,
    new_book_shelf
):
    """
    Добавляет новую книгу в библиотеку и на указанную полку.

    Проверяет существование полки. Если полка существует, создаёт словарь
    с данными о книге, добавляет его в общий список книг, а номер книги —
    в список на полке. Если полки не существует, возвращает сообщение об ошибке.

    Args:
        books (list): Список всех книг в библиотеке (каждая книга — словарь).
        directories (dict): Словарь полок, где ключ — номер полки,
                            значение — список номеров книг на этой полке.
        new_book_number (str): Уникальный номер добавляемой книги.
        new_book_genre (str): Жанр книги.
        new_book_title (str): Название книги.
        new_book_author (str): Автор книги.
        new_book_shelf (str): Номер полки, на которую нужно поместить книгу.

    Returns:
        str: Сообщение о результате операции (успех или ошибка).
    """
    if new_book_shelf in directories:
        new_book = {
            'number': new_book_number,
            'genre': new_book_genre,
            'title': new_book_title,
            'author': new_book_author
        }
        books.append(new_book)
        directories[new_book_shelf].append(new_book_number)
        return 'Книга добавлена. Текущий список книг:'
    else:
        return 'Такой полки не существует. Добавьте полку командой add_shelf.\nТекущий список книг:'

def add_new_book(books, directories):
    """
    Обрабатывает команду добавления новой книги.

    Запрашивает у пользователя данные о книге (номер, жанр, название, автора)
    и номер полки для хранения. Вызывает функцию add_new_book_to_library()
    для выполнения операции и выводит результат. Затем отображает актуальный
    список всех книг.

    Args:
        books (list): Список всех книг в библиотеке.
        directories (dict): Словарь полок с номерами книг.

    Returns:
        None
    """
    new_book_number = input('Введите номер книги:\n')
    new_book_genre = input('Введите жанр книги:\n')
    new_book_title = input('Введите название книги:\n')
    new_book_author = input('Введите автора книги:\n')
    new_book_shelf = input('Введите полку для хранения:\n')
    result = add_new_book_to_library(
        books,
        directories,
        new_book_number,
        new_book_genre,
        new_book_title,
        new_book_author,
        new_book_shelf
    )
    print(result)
    all_(books, directories)
````

## **Основная программа**

````python
def main(books, directories):
    """
    Запускает основной цикл программы для обработки пользовательских команд.
    Пользователь может вводить команды для получения информации о книгах,
    работы с полками и просмотра данных. Выполнение продолжается до ввода
    команды "q", завершающей программу.
    Args:
        books (list): Список словарей с данными о книгах.
        directories (dict): Словарь с полками и списками номеров книг.
    Returns:
        None
    """
    while True:
        command = input('Введите команду (book_info, shelf, all, add_shelf, del_shelf, add_book, q):\n')
        print()

        match command:
            case 'book_info':
                book_info(books)
            case 'shelf':
                shelf(books, directories)
            case 'all':
                all_(books, directories)
            case 'add_shelf':
                add_shelf(directories)
            case 'del_shelf':
                del_shelf(directories)
            case 'add_book':
                add_new_book(books, directories)
            case 'q':
                print('Конец программы')
                break
            case _:
                print('Неизвестная команда, введите корректную')

main(books, directories)
````
