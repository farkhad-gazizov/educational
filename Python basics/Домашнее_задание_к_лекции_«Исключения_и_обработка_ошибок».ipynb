**Задание 2**

Дан поток дат в формате YYYY-MM-DD, в которых встречаются некорректные значения:

`stream = [‘2018-04-02’, ‘2018-02-29’, ‘2018-19-02’]`

Напишите функцию, которая проверяет эти даты на корректность. То есть для каждой даты возвращает True — дата корректна или False — некорректная.

````python
from datetime import datetime

stream = ['2018-04-02', '2018-02-29', '2018-19-02']

def date_is_correct(stream):
    for date_string in stream:
        try:
            # перевод каждой строки в потоке в объект datetime
            datetime.strptime(date_string, '%Y-%m-%d')
            # вывод True, если нет ошибки, то есть, дата корректная
            print(f'{date_string} - True')
        except ValueError:
            # обработка некорректных дат и вывод False
            print(f'{date_string} - False')

date_is_correct(stream)
````

**Задание 3**

Напишите функцию `date_range`, которая возвращает список дат за период от `start_date` до `end_date`. Даты должны вводиться в формате `YYYY-MM-DD`. В случае неверного формата или при `start_date > end_date` должен возвращаться пустой список.

````python
from datetime import datetime, timedelta

start_date = '2026-05-01'
end_date = '2026-05-09'

def date_range(start_date, end_date):
    dates = []
    try:
        '''
        начальному значению текущей даты (объект datetime),
        которая будет перебираться и увеличиваться в цикле while,
        присваиваем значение стартового дня
        '''
        current_date_dt = datetime.strptime(start_date, '%Y-%m-%d')
        end_date_dt = datetime.strptime(end_date, '%Y-%m-%d')

        while current_date_dt <= end_date_dt:
            '''
            пополнение списка дат dates по их возрастанию при верном
            формате дат и при start_date <= end_date
            '''
            dates.append(current_date_dt.strftime('%Y-%m-%d'))
            current_date_dt += timedelta(days=1)

    except ValueError:
        pass
        '''
        список дат dates останется пустым в случае неверного формата дат
        и при start_date > end_date (т.е. ошибок формата даты), при этом другие
        некасающиеся условия задачи ошибки будут выведены
        '''

    return dates

print(date_range(start_date, end_date))
````
